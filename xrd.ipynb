```python
#importation de package
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.analysis.diffraction.core import AbstractDiffractionPatternCalculator
```

```python
#Structure du ScCuS2
with MPRester("RPlDkSQDkfldGfeTdTo") as m:

    # Structure for material id
    s = m.get_structure_by_material_id("mp-6980")
```

```python
#pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å
self = XRDCalculator('CuKa')
pattern = self.get_pattern(s)
```

```python
#Coordonnées du graphe de diffraction
print(pattern)
```

```python
plot = AbstractDiffractionPatternCalculator.get_plot(self,s,fontsize=10)
```
