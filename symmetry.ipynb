```python
#importation de package
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.symmetry.analyzer import PointGroupAnalyzer

import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
#Structure du ScCuS2
with MPRester("RPlDkSQDkfldGfeTdTo") as m:

    # Structure for material id
    s = m.get_structure_by_material_id("mp-6980")
```

```python
#liste des opérations de symmétrie
symm_op = SpacegroupAnalyzer(s).get_symmetry_operations()
```

```python
print(symm_op)
```

```python
#coordonnées des atomes de ScCuS2 trouvées avec Jmol Crystal Symmetry Explorer
Sc_0 = [0,0,0]
Cu_1 = [0.33333,0.66667,0.40]
S_2 = [0.33333,0.66667,0.78]

print('Coordonnées initiales')
print('Sc_0 : ')
print(Sc_0)
print('Cu_1 : ')
print(Cu_1)
print('S_2 : ')
print(S_2)
```

```python
#Première operation (-y,x-y,z) sur Sc_0
op1 = symm_op[1].operate(Sc_0)
#Deuxième operation (-x+y,-x,z) sur Cu_1
op2 = symm_op[2].operate(Cu_1)
#Troisième operation (-y,-x,z) sur S_2
op3 = symm_op[3].operate(S_2)

print('Coordonnées finales')
print('Sc_0 : ')
print(symm_op[1].rotation_matrix,'*',Sc_0,'+',symm_op[1].translation_vector,'=',op1)
print('Cu_1 : ')
print(symm_op[2].rotation_matrix,'*',Cu_1,'+',symm_op[2].translation_vector,'=',op2)
print('S_2 : ')
print(symm_op[3].rotation_matrix,'*',S_2,'+',symm_op[3].translation_vector,'=',op3)
```

```python
print('Opération (-y,x-y,z) sur Sc_0')

Sc_0_op1 = img.imread('Sc_0_op1.png')
fig, axes = plt.subplots(figsize=(10,10))
axes.imshow(Sc_0_op1)
plt.show()
```

```python
print('Opération (-x+y,-x,z) sur Cu_1')

Cu_1_op2 = img.imread('Cu_1_op2.png')
fig, axes = plt.subplots(figsize=(10,10))
axes.imshow(Cu_1_op2)
plt.show()
```
